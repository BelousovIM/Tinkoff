```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import datetime

orange = plt.cm.Oranges(0.5)
green = plt.cm.Greens(0.5)

df = pd.read_csv('dz-z2.csv',sep=',', na_values='NaN')
df['receive_dt']=pd.to_datetime(df['receive_dt'],format='%d%b%Y')
df['Дата'] = df['receive_dt'].dt.strftime('%Y/%m/%d')
df['use_dt']=pd.to_datetime(df['use_dt'],format='%d%b%Y')
df['use_dt'] = df['use_dt'].dt.strftime('%Y/%m/%d')




df1=df[df.card_nm=='Alpha'].groupby('Дата')[['card_id']].count()
df1['her']=df[(df.card_nm=='Alpha')&(df.use_dt!='NaT')].groupby('Дата')[['card_id']].count()    
df1['card_sum']=df1['her_sum']=158*[0]
A=[]
B=[]
A.append(df1['card_id'].head(1).sum())
B.append(df1['her'].head(1).sum())
for k in range(2,159):
    A.append(df1['card_id'].head(k).sum())
    B.append(df1['her'].head(k).sum())
df1['card_sum']=A
df1['her_sum']=B
A.clear;B.clear
df1['kon']=df1['her_sum']/df1['card_sum']*100

df2=df[df.card_nm=='Beta'].groupby('Дата')[['card_id']].count()
df2['her']=df[(df.card_nm=='Beta')&(df.use_dt!='NaT')].groupby('Дата')[['card_id']].count()    
df2['card_sum']=df2['her_sum']=158*[0]
A=[]
B=[]
A.append(df2['card_id'].head(1).sum())
B.append(df2['her'].head(1).sum())
for k in range(2,159):
    A.append(df2['card_id'].head(k).sum())
    B.append(df2['her'].head(k).sum())
df2['card_sum']=A
df2['her_sum']=B
A.clear;B.clear
df2['kon']=df2['her_sum']/df2['card_sum']*100

df3=df[df.card_nm=='Gamma'].groupby('Дата')[['card_id']].count()
df3['her']=df[(df.card_nm=='Gamma')&(df.use_dt!='NaT')].groupby('Дата')[['card_id']].count()    
df3['card_sum']=df3['her_sum']=158*[0]
A=[]
B=[]
A.append(df3['card_id'].head(1).sum())
B.append(df3['her'].head(1).sum())
for k in range(2,159):
    A.append(df3['card_id'].head(k).sum())
    B.append(df3['her'].head(k).sum())
df3['card_sum']=A
df3['her_sum']=B
A.clear;B.clear
df3['kon']=df3['her_sum']/df3['card_sum']*100

fig, ax = plt.subplots(figsize=(32, 12))
ax.set_xlabel(' ',fontsize=35)
ax.set_ylabel('Конверсия %',fontsize=30)
ax.tick_params( axis='y',labelsize=25)
ax.tick_params( axis='x',labelsize=25)
df1['kon'].plot(figsize=(37, 17), ax=ax).set_ylim(0,60);
df2['kon'].plot(ax=ax, color=orange);
df3['kon'].plot(ax=ax, color=green);
ax.legend(["Alpha", "Beta",'Gamma'],loc=1, handlelength=5,borderpad=1,prop={'size':25});
```
